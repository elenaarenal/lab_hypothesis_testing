# Challenge 2 - Hypothesis Tests

In this section of the lab, we will test whether the hourly wage of all hourly workers is significantly different from $30/hr. Import the correct one sample test function from scipy and perform the hypothesis test for a 95% two sided confidence interval.

```python
# Your code here:
db_hour = db[(db['Salary or Hourly'] == ('Hourly'))]
db_hour.dropna(subset=['Salary or Hourly'], inplace=True)

db_hour['total']=db_hour['Typical Hours']*db_hour["Hourly Rate"]

db_hour['total'].mean()

print('mean hourly', db_hour['Hourly Rate'].mean())
```

```python
# Test parameters 

hypothesis_mean = 30

sample_mean = db_hour['Hourly Rate'].mean()

alpha = 0.05

print(f'Hypothesis mean: {hypothesis_mean} USD',
      f'\nSample mean: {sample_mean} USD',
      f'\nProbability threshold: {alpha}')
```

```python
db_hour.head()
```

```python

t_stat, p_value = stats.ttest_1samp(db_hour['Hourly Rate'], 30)
print(t_stat)
print(p_value)
#t<p: True
```

```python
# Z-test (the probability that the sample mean is not within the 95% of the population mean) Otra manera

Z_score, p_value = ztest(db_hour['Hourly Rate'], value=hypothesis_mean)

print(f'Z_score: {Z_score}', f'\np-value: {p_value}')

if p_value < alpha:
    print("Salary per hour is significantly different to $30/h.")
else:
    print("DO not reject the null hypotesis, No rechazamos la hipÃ³tesis nula: we can say that the salary per hour is not significantly different from $30/h.")
```

# Challenge 3: To practice - Constructing Confidence Intervals

While testing our hypothesis is a great way to gather empirical evidence for accepting or rejecting the hypothesis, another way to gather evidence is by creating a confidence interval. A confidence interval gives us information about the true mean of the population. So for a 95% confidence interval, we are 95% sure that the mean of the population is within the confidence interval. 
).

To read more about confidence intervals, click [here](https://en.wikipedia.org/wiki/Confidence_interval).


In the cell below, we will construct a 95% confidence interval for the mean hourly wage of all hourly workers. 

The confidence interval is computed in SciPy using the `t.interval` function. You can read more about this function [here](https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.t.html).

To compute the confidence interval of the hourly wage, use the 0.95 for the confidence level, number of rows - 1 for degrees of freedom, the mean of the sample for the location parameter and the standard error for the scale. The standard error can be computed using [this](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.sem.html) function in SciPy.

```python
# Your code here:

# Calculate mean and standard deviation
mean_wage = db_hour['Hourly Rate'].mean()
std_dev = db_hour['Hourly Rate'].std()

# Number of samples
n = len(db_hour)

# Calculate standard error of the mean
std_error = std_dev / np.sqrt(n)

# Degrees of freedom
df = n - 1

# Confidence level
confidence_level = 0.95

# Calculate confidence interval using t distribution
confidence_interval = t.interval(confidence_level, df, loc=mean_wage, scale=std_error)

print(f"Mean Hourly Wage: {mean_wage}")
print(f"Standard Error of the Mean: {std_error}")
print(f"95% Confidence Interval for the Mean Hourly Wage: {confidence_interval}")
```

Now construct the 95% confidence interval for all salaried employeed in the police in the cell below.

```python
# Your code here:

# Calculate mean and standard deviation
mean_salary = db_polis['Annual Salary'].mean()
std_dev = db_polis['Annual Salary'].std()

# Number of samples
n = len(db_polis)

# Calculate standard error of the mean
std_error = std_dev / np.sqrt(n)

# Degrees of freedom
df = n - 1

# Confidence level
confidence_level = 0.95

# Calculate confidence interval using t distribution
confidence_interval = t.interval(confidence_level, df, loc=mean_salary, scale=std_error)

print(f"Average: {mean_salary}")
print(f"Standard Error of the Average(mean): {std_error}")
print(f"95% Confidence Interval for the Salary : {confidence_interval}")
```
